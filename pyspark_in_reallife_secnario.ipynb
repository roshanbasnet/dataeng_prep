```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("spark_session1").getOrCreate()
```

```python
df_emp = spark.read.option("nullValue","null").csv("emp.csv",header=True, inferSchema=True)
df_emp.show()
```

```python
from pyspark.sql.functions import to_date

#Change string to Date DataType

df_emp = df_emp.withColumn("HIREDATE",to_date("HIREDATE","dd-MM-yyyy")).fillna({"HIREDATE":'9999-12-31'})
df_emp.show()
```

```python
from pyspark.sql.functions import date_format
#creating two YEAR and Month new column based on year
df_emp = df_emp.withColumn("YEAR",date_format("HIREDATE","yyyy")).withColumn("MONTH",date_format("HIREDATE","MM"))
df_emp.show()
```
